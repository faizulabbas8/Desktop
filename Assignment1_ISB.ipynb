```python
# Load each sheet into a separate DataFrame
df1 = pd.read_excel(output, sheet_name="Dataset")
df2 = pd.read_excel(output, sheet_name="Payment Type")
df3 = pd.read_excel(output, sheet_name="Employment Type")
```

```python
# convert values prev_address_months_count to +ve
df1['prev_address_months_count'] = df1['prev_address_months_count'].abs()
```

```python
#Just Checking it's working for 1 column let's say income
df1['income'] = pd.to_numeric(df1['income'], errors='coerce')
```

```python
# Select columns with numeric data types
numeric_columns = df1.select_dtypes(include=['number']).columns
```

```python
# convert all the numeric column which has string as numeric & convert numeric columns to numeric type, handling errors
for columns in numeric_columns:
  df1[columns] = pd.to_numeric(df1[columns], errors='coerce')
```

```python
# Create a dictionary for mapping payment types
payment_type_mapping = df2.set_index('payment_type')['Name'].to_dict()

# Create a dictionary for mapping employment statuses
employment_type_mapping = df3.set_index('employment_status')['Name'].to_dict()

# Replace values in df1
df1['payment_type'] = df1['payment_type'].map(payment_type_mapping)
df1['employment_status'] = df1['employment_status'].map(employment_type_mapping)

# Display the updated df1
df1.head(10)
```
